```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
loan_dataset = pd.read_csv('/content/train_u6lujuX_CVtuZ9i (1).csv')
```

```python
loan_dataset.dropna(inplace=True)
```

```python
loan_dataset.Dependents.unique()
```

```python
loan_dataset.Property_Area.unique()
```

```python
loan_dataset.replace({'Loan_Status':{'N':0, 'Y':1}}, inplace=True)
```

```python
loan_dataset.replace({'Married':{'No':0, 'Yes':1}}, inplace=True)
```

```python
loan_dataset.replace({'Self_Employed':{'No':0, 'Yes':1}}, inplace=True)
```

```python
loan_dataset
```

```python
loan_dataset.replace({'Dependents':{'4':4, '0':0, '1':1, '2':2}}, inplace=True)
```

```python
loan_dataset.replace({'Education':{'Graduate':1, 'Not Graduate':0}}, inplace=True)
```

```python
loan_dataset.replace({'Gender':{'Male':1, 'Female':0}}, inplace=True)
```

```python
loan_dataset.info()
```

```python
loan_dataset.head()
```

```python
loan_dataset.replace({'Property_Area':{'Rural':0, 'Semiurban':1, 'Urban':2}}, inplace=True)
```

```python
x = loan_dataset.drop(columns=['Loan_ID', 'Loan_Status'], axis=1)
y = loan_dataset['Loan_Status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, stratify=y, random_state=2)
```

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(x_train, y_train)
```

```python
x_train_prediction = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
training_data_accuracy
```

```python
x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
test_data_accuracy
```
